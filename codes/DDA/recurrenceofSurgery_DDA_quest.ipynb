```python
Q1 = df_ROS_drops['연령'].describe()['25%']
Q3 = df_ROS_drops['연령'].describe()['75%']
IQR = Q3 - Q1
```

```python
bound_up = Q3 + 1.5 * IQR
bound_low = Q1 - 1.5 * IQR
bound_low,bound_up
```

```python
condition = (df_ROS_drops['연령'] <= bound_up) #긍정 조건 변환
condition
```

```python
df_ROS_condition = df_ROS_drops[condition]
```
